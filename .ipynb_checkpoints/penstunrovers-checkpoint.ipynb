````python
def getSeason(S):
#     gameList = []
    sDF = pd.DataFrame()
    for p in next(os.walk('Boxscores/S%s/'%S))[1]:
        if p[0] != '.':
            teamStats = pd.read_csv('Boxscores/S%s/%s/TeamStats.csv'%(S,p),index_col=1)
            teamStats = teamStats.loc[:,~teamStats.columns.str.contains('^Unnamed')]
            box = pd.read_csv('Boxscores/S%s/%s/Boxscore.csv'%(S,p)).transpose()
            box.columns = teamStats.columns
            game = pd.concat([teamStats,box])
            if game.loc['F'].iloc[0] == game.loc['F'].iloc[1]:
                continue
            elif game.loc['F'].iloc[0] > game.loc['F'].iloc[1]:
                gameDF = pd.DataFrame(game.iloc[:,0]>game.iloc[:,1],columns=[p]).transpose()
            else:
                gameDF = pd.DataFrame(game.iloc[:,0]<game.iloc[:,1],columns=[p]).transpose()
            gameDF['S'] = int(S)
            gameDF.index.name = 'gameID'
            gameDF = gameDF.reset_index().set_index(['S','gameID'])
            sDF = pd.concat([sDF,gameDF])
    return sDF.sort_values(['S','gameID'])
````

````python
curS = 26
````

````python
allDF = pd.DataFrame()
sBar = tqdm(total=curS)
for S in range(1,curS+1):
    sDF = getSeason(S)
    allDF = pd.concat([allDF, sDF])
    sBar.update(1)
````
