```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from tqdm.notebook import tqdm
```

```python
def getSeasonStats(S):
    # S=25
    rushDF = pd.concat([pd.read_csv('Boxscores/S%s/%s/rushStats.csv'%(S,p)) for p in next(os.walk('Boxscores/S%s/'%S))[1] if p[0] != '.'])
    recDF = pd.concat([pd.read_csv('Boxscores/S%s/%s/recStats.csv'%(S,p)) for p in next(os.walk('Boxscores/S%s/'%S))[1] if p[0] != '.'])
    specDF = pd.concat([pd.read_csv('Boxscores/S%s/%s/specStats.csv'%(S,p)) for p in next(os.walk('Boxscores/S%s/'%S))[1] if p[0] != '.'])
    puntDF = pd.concat([pd.read_csv('Boxscores/S%s/%s/puntStats.csv'%(S,p)) for p in next(os.walk('Boxscores/S%s/'%S))[1] if p[0] != '.'])

    kickDF = pd.concat([pd.read_csv('Boxscores/S%s/%s/kickStats.csv'%(S,p)) for p in next(os.walk('Boxscores/S%s/'%S))[1] if p[0] != '.'])
#     kickList = ['FG < 20','20-29','30-39','40-49','50+']
#     for k in kickList:
#         kickDF['%sM'%k] = [int(x.strip('[]').split(',')[0]) for x in kickDF[k]]
#         kickDF['%sA'%k] = [int(x.strip('[]').split(',')[1]) for x in kickDF[k]]
#     kickDF = kickDF.drop(columns=kickList)

    defDF = pd.concat([pd.read_csv('Boxscores/S%s/%s/defStats.csv'%(S,p)) for p in next(os.walk('Boxscores/S%s/'%S))[1] if p[0] != '.'])
#     fumList = [int(x.strip('[]').split('/')[0]) for x in defDF['FF/FR']]
#     defDF['FF'] = fumList[0]
#     defDF['FR'] = fumList[1]
#     blkList = [int(x.strip('[]').split(',')[0]) for x in defDF['Blk P/XP/FG']]
#     defDF['Blk P'] = blkList[0]
#     defDF['Blk XP'] = blkList[1]
#     defDF['Blk FG'] = blkList[2]
#     defDF = defDF.drop(columns=['FF/FR','Blk P/XP/FG'])

    passDF = pd.concat([pd.read_csv('Boxscores/S%s/%s/passStats.csv'%(S,p)) for p in next(os.walk('Boxscores/S%s/'%S))[1] if p[0] != '.'])
#     cpatList = [int(x.strip('[]').split(',')[0]) for x in passDF['Cp/At']]
#     passDF['Cmp'] = cpatList[0]
#     passDF['Att'] = cpatList[1]
#     passDF = passDF.drop(columns=['Cp/At'])

    for df in [passDF,rushDF,recDF,specDF,puntDF,kickDF,defDF]:
        df['S'] = str(S)
        df['Team'] = df['Team'].str.replace('ARI','AZ')
        
    return [passDF,rushDF,recDF,specDF,puntDF,kickDF,defDF]
```

```python
passList = []
rushList = []
recList = []
defList = []
specList = []
kickList = []
puntList = []

curSeason = 25
sList = range(1,curSeason+1)

sBar = tqdm(total=len(sList))
for s in sList:
    seasonList = getSeasonStats(s)
    passList.append(seasonList[0])
    rushList.append(seasonList[1])
    recList.append(seasonList[2])
    specList.append(seasonList[3])
    puntList.append(seasonList[4])
    kickList.append(seasonList[5])
    defList.append(seasonList[6])
    sBar.update(1)
```

```python
passDF = pd.concat(passList)
rushDF = pd.concat(rushList)
rushDF['RushYds'] = rushDF['Yds']
rushDF = rushDF.drop('Yds',axis=1)
recDF = pd.concat(recList)
recDF['RecYds'] = recDF['Yds']
recDF = recDF.drop('Yds',axis=1)
specDF = pd.concat(specList)
puntDF = pd.concat(puntList)
kickDF = pd.concat(kickList)
defDF = pd.concat(defList)
```
